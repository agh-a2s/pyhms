### Coverage visualization

```python
import matplotlib.pyplot as plt
import dill as pkl
```

```python
with open("results/simple_sprout.pkl", "rb") as data_file:
    s_far_enough, s_bench_deme_numbers, s_bench_coverages, s_bench_optima_numbers = pkl.load(data_file)
with open("results/nbc_cut.pkl", "rb") as data_file:
    _, c_bench_deme_numbers, c_bench_coverages, c_bench_optima_numbers = pkl.load(data_file)
    c_cut = np.linspace(1.0, 3.0, 21)
with open("results/nbc_far_enough.pkl", "rb") as data_file:
    _, f_bench_deme_numbers, f_bench_coverages, f_bench_optima_numbers = pkl.load(data_file)
    f_far_enough = np.linspace(1.0, 3.0, 21)
with open("results/nbc_trunc.pkl", "rb") as data_file:
    _, t_bench_deme_numbers, t_bench_coverages, t_bench_optima_numbers = pkl.load(data_file)
    t_trunc = np.linspace(0.1, 0.91, 17)
```

```python
rows = 4
cols = 3
col_to_bench = {0: 4, 1: 7, 2: 9}
fig, axes = plt.subplots(rows, cols, figsize=(10, 14), sharey=True)
fig.subplots_adjust(wspace=0, hspace=0.3)
for col in range(cols):
    axes[0, col].set_title("Target function {}".format(col_to_bench[col]+1))
    axes[0, col].plot(s_far_enough, s_bench_deme_numbers[col_to_bench[col]], label="Subprocess count", linestyle='dashed')
    axes[0, col].plot(s_far_enough, s_bench_coverages[col_to_bench[col]], label="Number of optima covered", linestyle='solid')
    axes[0, col].axhline(y = s_bench_optima_numbers[col_to_bench[col]], color = 'black', linestyle = 'dotted', label='Total number of optima')
    axes[0, col].yaxis.set_ticks(np.arange(0.0, 21.0, 2.0))
    axes[1, col].plot(c_cut, c_bench_deme_numbers[col_to_bench[col]], label="Subprocess count", linestyle='dashed')
    axes[1, col].plot(c_cut, c_bench_coverages[col_to_bench[col]], label="Number of optima covered", linestyle='solid')
    axes[1, col].axhline(y = c_bench_optima_numbers[col_to_bench[col]], color = 'black', linestyle = 'dotted', label='Total number of optima')
    axes[1, col].yaxis.set_ticks(np.arange(0.0, 21.0, 2.0))
    axes[2, col].plot(f_far_enough, f_bench_deme_numbers[col_to_bench[col]], label="Subprocess count", linestyle='dashed')
    axes[2, col].plot(f_far_enough, f_bench_coverages[col_to_bench[col]], label="Number of optima covered", linestyle='solid')
    axes[2, col].axhline(y = f_bench_optima_numbers[col_to_bench[col]], color = 'black', linestyle = 'dotted', label='Total number of optima')
    axes[2, col].yaxis.set_ticks(np.arange(0.0, 21.0, 2.0))
    axes[3, col].plot(t_trunc, t_bench_deme_numbers[col_to_bench[col]], label="Subprocess count", linestyle='dashed')
    axes[3, col].plot(t_trunc, t_bench_coverages[col_to_bench[col]], label="Number of optima covered", linestyle='solid')
    axes[3, col].axhline(y = t_bench_optima_numbers[col_to_bench[col]], color = 'black', linestyle = 'dotted', label='Total number of optima')
    axes[3, col].yaxis.set_ticks(np.arange(0.0, 21.0, 2.0))
    if col == 1:
        axes[0, col].set_xlabel('Value of far enough parameter')
        axes[1, col].set_xlabel('Value NBC cut parameter')
        axes[2, col].set_xlabel('Value NBC far enough parameter')
        axes[3, col].set_xlabel('Value NBC truncation parameter')
handles, labels = axes[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center")
plt.show()
```
